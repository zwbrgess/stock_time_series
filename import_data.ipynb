```python
import yfinance as yf

start_date = "2020-01-01"
end_date = "2023-10-24"
df = yf.download('AAPL', start=start_date, end=end_date)
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt

df['Close'].plot(title='AAPL Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

```python
df['Volume'].plot(title='AAPL Trading Volume', color='gray')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
df['50-day MA'] = df['Close'].rolling(50).mean()
df['200-day MA'] = df['Close'].rolling(200).mean()

df[['Close', '50-day MA', '200-day MA']].plot(title='AAPL with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```
